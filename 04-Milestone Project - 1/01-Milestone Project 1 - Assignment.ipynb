```python
def display_game(x_positions, o_positions):
    # displays game baord
    # Inputs: list of x positions, list of y positions
    # Ouput: Prints board 
    # replace an empty character with x or y

    board = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']

    if not x_positions and not o_positions:
        # both are empty, print empty board
        board = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    # if x or o is not empty, subtract 1 
    elif x_positions != [] and not o_positions:
        print(f"subtracting 1 from {x_positions}")
        x_positions = [pos-1 for pos in x_positions]
    else:
        print(f"subtracting 1 from x = {x_positions} and o = {o_positions}")
        x_positions = [pos-1 for pos in x_positions]
        o_positions = [pos-1 for pos in o_positions] 

    # replace empty board spaces with X and Os
    for idx in x_positions:
        board[idx] = 'X'
    for idx in o_positions: 
        board[idx] = 'O'

    print(board[0:3])
    print(board[3:6]) 
    print(board[6:])
```

```python
def user_choice(current_player, x_positions, o_positions, possible_values):
    # inputs: 
    # current_player: X or O
    # x_positions
    # y_positions
    # possible_values: list of available spaces to choose
    # outputs:
    # x_positions: list of x positions
    # o_positions: list of o positions

    choice = 'wrong'

    while choice not in possible_values:
        print(f"Current player is {current_player}")
        print(f"Possible values are {possible_values}")
        choice = input(f"Where would you like to place your {current_player}?")
        print(f"choice is {choice}")
        if choice not in possible_values:
            clear_output()
            print("Sorry, choose a valid position.")

    # Choice is now valid
    possible_values.remove(choice) # remove choice from string list of possible values
    choice = int(choice) # cast to integer

    # add choice to appropriate list
    if current_player == 'X':
        x_positions.append(choice)
    elif current_player == 'O':
        o_positions.append(choice)

    # return Xs, Os, and remaining possible board values
    return x_positions, o_positions, possible_values

```

```python
def win_check(x_positions, o_positions, possible_values, game_on): 
    # checks list of positions and sees if there's a winner
    
    win_states = [{1,2,3}, {4,5,6}, {7,8,9}, # horizontal
                  {1,4,7}, {2,5,8}, {3,6,9}, # vertical
                  {1,5,9}, {3,5,7}] # diagonal
    
    print(f"inside win check possible values are {possible_values}")
    #compare x_positions and o_positions sets to see if they match 
    print(f"checking x = {set(x_positions)} wins?")
    print(f"checking o = {set(o_positions)} wins?")
    
    for winning_sets in win_states:
        if set(x_positions).issuperset(winning_sets): # Check if x wins
            print("X wins!")
            game_on = False # end game
        elif set(o_positions).issuperset(winning_sets): # Check if o wins
            print("O wins!")
            game_on = False
    #else: # do nothing if no winner
        #pass 
    # return game on state and print out winner
    
    print(f"inside win check possible values are {possible_values}")

    if game_on == False:
        # ask if user wants to continue playing
        choice = 'wrong'

        while choice not in ['y', 'n']:
            choice = input("Would you like to restart the game? [y/n]")

            if choice.lower() not in  ['y', 'n']:
                clear_output()
                print("Wrong input, please type 'y' or 'n'")
        # return empty x and o if game win, otherwise return inputted x and o positions

        # choice now valid
        if choice == 'y':
            # Reset game state
            game_on = True
            x_positions = []
            o_positions = []
            possible_values = ['1','2','3','4','5','6','7','8','9']
        #else: # do nothing
          #  pass
   # else:
    #    pass
    
    print(f"inside win check possible values are {possible_values}")


    return game_on, x_positions, o_positions, possible_values

```

```python
def switch_player(current_player):
    # input current player
    # output next player

    if current_player == 'X':
        player = 'O'
    elif current_player =='O':
        player = 'X'
    
    return player
```
